```python
students_score_sheet=[]   #list

score1={'name:':'May','Listening:':'7.5','Reading:':'8.0','Writing:':'7.5','Speaking:':'8.5','Overall:':'8.0'}

score2={'name:':'Charlotte','Listening:':'5.5','Reading:':'5.0','Writing:':'5.5','Speaking:':'5.5','Overall:':'5.5'}

#dictionary

students_score_sheet.append(score1)
students_score_sheet.append(score2)

for student in students_score_sheet:

    for key in student:
     print(key,student[key])
    print('==============')

```

```python
import random
students_score_sheet=[]   #list

for x in range(10):
    scores={} #dict 
    scores['name']=f'學生{x+1}'
    scores['Listening']= random.randint(0.0,9.0)
    scores['Reading']= random.randint(0.0,9.0)
    scores['Writing']= random.randint(0.0,9.0)
    scores['Speaking']= random.randint(0.0,9.0)
    scores['Overall']= random.randint(0.0,9.0)
    students_score_sheet.append(scores)

print (students_score_sheet)

for scores in students_score_sheet:

  for key in scores:
     print(key, scores[key])
  print('==============')
     
```
